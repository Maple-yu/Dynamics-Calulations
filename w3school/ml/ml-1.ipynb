## regression
### 1.Line Regression

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

### using lineregress to get the returned tuple of parameter 
slope, intercept, r, p, std_err = stats.linregress(x, y)
### the degree of correlation
print(r)

def myfunc(x):
  return slope * x + intercept
speed = myfunc(10)
print('the speed of predict :',speed)
mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

### 2.Polynomial Regression

```python
### r2_score : measure the degree of correlatio
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

myline = numpy.linspace(1, 22, 100)
### show how it fits
print(r2_score(y, mymodel(x)))

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

### 3. Multiple Regression

```python
import pandas
from sklearn import linear_model

df = pandas.read_csv("./data/cars.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

### Predict the CO2 of the car with 2300kg and 1300ccm 
predictedCO2 = regr.predict([[2300, 1300]])
print('The predicted CO2 :',predictedCO2)
predictedCO2 = regr.predict([[3300, 1300]])
print('The predicted CO2 of add 1000g:',predictedCO2)

### Reduce a degree，get the walue when Weight add 1 or Volume added 1
print('Weight add 1 or Volume added 1',regr.coef_)
```
