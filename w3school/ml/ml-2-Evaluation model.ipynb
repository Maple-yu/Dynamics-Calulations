## Training / testing
To measure whether the model is good enough, we can use a method called training/testing
### 1.database

```python
import numpy
import matplotlib.pyplot as plt
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

### split into the train and test
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

plt.scatter(train_x, train_y)
plt.show()
```

### 2. The fit database

```python
import numpy
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

numpy.random.seed(2)
x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))
### R-squared of the train database!
r2 = r2_score(train_y, mymodel(train_x))
print('R-squared of the train database',r2)

### R-squared of the test database.
r2 = r2_score(test_y, mymodel(test_x))
print('R-squared of the test database',r2)

myline = numpy.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()

### Let's predict it !
print('Money spent by a customer staying in the store for five minutes is predicted :',mymodel(5))
```
